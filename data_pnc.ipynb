```python
import pandas as pd
import datetime
from datetime import datetime
from pandas import DataFrame, Series
import numpy as np
```

```python
df = pd.read_excel(open('C:\Intern_data\pnc_data_analytic\data\Datapc11072019.xls','rb'),sheet_name= "Trienkhai")
 
df.head()
```

```python
time_d = ['Ngày hoàn tất','checkindateFDX','checkoutdateFDX','checkindateFtel','checkoutdateFTEL','Nhận tuyến','Nhả Tuyến','Thời gian phân công','Ngày hẹn','Ngày tạo giao dịch']
def func_conver_gio2(S):
    if pd.notnull(S):
        S = S[:-16] + S[len(S) - 2:len(S)]
        x = datetime.strptime(str(S), '%d-%b-%y %I.%M%p')
    else:
        x = pd.NaT
    return x
for x in time_d:
    df[x] = df[x].apply(func_conver_gio2)
    
df.head()
```

```python
data = df[['Blockid','Số hợp đồng','Emp_id','Ngày tạo giao dịch','Thời gian phân công','Ngày hẹn','Nhận tuyến','Nhả Tuyến','checkindateFDX','checkoutdateFDX','checkindateFtel','checkoutdateFTEL']]
data.info()
```

```python
data = data.rename({'Số hợp đồng':'So_hd','Ngày tạo giao dịch':'ngay_tao','Thời gian phân công':'thoi_gian_phan_cong',
                        'Ngày hẹn':'ngay_hen','Nhận tuyến':'nhan_tuyen','Nhả Tuyến':'nha_tuyen','checkindateFtel':'checkindateFTEL'}, axis = 'columns')
data.head()
```

```python
data['duration1']= data['ngay_hen']-data['ngay_tao']
data['duration1'] = data['duration1'].dt.days*24 +data.duration1.dt.seconds/3600

data['duration2']= data['thoi_gian_phan_cong']-data['ngay_tao']
data['duration2'] = data['duration2'].dt.days*24 +data.duration2.dt.seconds/3600

data['duration3']= data['checkindateFDX']-data['thoi_gian_phan_cong']
data['duration3'] = data['duration3'].dt.days*24 +data.duration3.dt.seconds/3600

data['duration4']= data['checkoutdateFDX']-data['checkindateFDX']
data['duration4'] = data['duration4'].dt.days*24 +data.duration4.dt.seconds/3600

data['duration5']= data['checkindateFTEL']-data['checkindateFDX']
data['duration5'] = data['duration5'].dt.days*24 +data.duration5.dt.seconds/3600

data['duration6']= data['checkoutdateFTEL']-data['checkoutdateFDX']
data['duration6'] = data['duration6'].dt.days*24 +data.duration6.dt.seconds/3600

data['duration7']= data['checkoutdateFTEL']-data['checkindateFTEL']
data['duration7'] = data['duration7'].dt.days*24 +data.duration7.dt.seconds/3600

```

```python
# data1: dữ liệu khái quát khoảng thời gian từ khi tạo đến hẹn của khách hàng.
data1_or = data[['ngay_tao','ngay_hen','duration1']]

#làm sạch data1
data1 = data1_or.drop([1435,1980,673])
data1.sort_values('duration1')
data1.reindex(np.arange(2432))
```

```python
#Tạo cột mốc tạo và cột mốc hẹn
data1['moc_tao'] = data1['ngay_tao'].dt.hour
data1['moc_hen'] = data1['ngay_hen'].dt.hour
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1_mt = data1.groupby('moc_tao').count()
moc_tao = data1_mt['ngay_tao'].index
```

```python
plt.figure(figsize = (12,7))
x=[]
y=[]
for i,j in zip( moc_tao,data1_mt['ngay_tao']):
    for k in np.arange(24):
        if k not in moc_tao:
            y +=[0]
            x +=[k]
        else:
            y+= [j]
            x+=[i]
        
ax = sns.barplot(x = x, y = y)
plt.title("Biểu đồ phân bổ sô lượng theo móc thời gian tạo", size = 17)
plt.xlabel('Mốc tạo', size = 13)
plt.ylabel('Số lượng',size =13)

for p in ax.patches:
    if p.get_height() > 0:
        form = p.get_height()
    else:
        form =''
    ax.annotate(format(form), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

plt.show()
```
